```python
import matplotlib.pyplot as plt
from tensorflow.keras.utils import image_dataset_from_directory

# Parametreler
batch_size = 32
img_size = (224, 224)  # Görselleri bu boyuta yeniden ölçekle

# Eğitim ve doğrulama veri setlerini oluştur
train_dataset = image_dataset_from_directory(data_dir,
                                             validation_split=0.2,
                                             subset="training",
                                             seed=123,
                                             image_size=img_size,
                                             batch_size=batch_size)

val_dataset = image_dataset_from_directory(data_dir,
                                           validation_split=0.2,
                                           subset="validation",
                                           seed=123,
                                           image_size=img_size,
                                           batch_size=batch_size)

print("Eğitim ve doğrulama veri setleri başarıyla oluşturuldu.")

# Çiçek türlerini burada al
class_names = train_dataset.class_names  # Çiçek türleri

AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.cache().prefetch(buffer_size=AUTOTUNE)

print("Veri seti önbelleğe alındı ve optimize edildi.")

plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):  # Bir partideki verileri al
    for i in range(9):  # İlk 9 görseli çizdir
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]]) # class_names artık burada kullanılabilir
        plt.axis("off")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Modeli oluştur
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(224, 224, 3)),  # Görselleri 0-1 aralığına ölçekle
    layers.Conv2D(32, (3, 3), activation='relu'),  # İlk evrişim katmanı
    layers.MaxPooling2D((2, 2)),                  # İlk havuzlama katmanı
    layers.Conv2D(64, (3, 3), activation='relu'), # İkinci evrişim katmanı
    layers.MaxPooling2D((2, 2)),                  # İkinci havuzlama katmanı
    layers.Conv2D(128, (3, 3), activation='relu'),# Üçüncü evrişim katmanı
    layers.MaxPooling2D((2, 2)),                  # Üçüncü havuzlama katmanı
    layers.Flatten(),                             # Düzleştirme katmanı
    layers.Dense(128, activation='relu'),         # Tam bağlantılı katman
    layers.Dense(5, activation='softmax')         # Çıkış katmanı (5 sınıf için softmax)
])

# Modelin özetini yazdır
model.summary()
```

```python
model.compile(
    optimizer='adam',                        # Modeli optimize etmek için Adam kullanıyoruz
    loss='sparse_categorical_crossentropy',  # Çok sınıflı sınıflandırma için uygun kayıp fonksiyonu
    metrics=['accuracy']                     # Doğruluk metriği
)
```

```python
# Modeli eğit
history = model.fit(
    train_dataset,  # Eğitim veri seti
    validation_data=val_dataset,  # Doğrulama veri seti
    epochs=10  # Eğitim yapılacak iterasyon sayısı
)
```

```python
import matplotlib.pyplot as plt

# Eğitim ve doğrulama doğruluğunu çizin
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.xlabel('Epok')
plt.ylabel('Doğruluk')
plt.legend()
plt.title('Model Doğruluğu')
plt.show()

# Eğitim ve doğrulama kaybını çizin
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.xlabel('Epok')
plt.ylabel('Kayıp')
plt.legend()
plt.title('Model Kaybı')
plt.show()
```
